```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
sizes = [100, 1000, 10000, 100000, 1000000]


b_times_100 = [0.0000, 0.0000, 0.0000, 0.0000, 0.0010]
b_times_1000 = [0.0050, 0.0050, 0.0045, 0.0055, 0.0051]
b_times_10000 = [0.0738, 0.0951, 1.1093, 0.1054, 1.1090]
b_times_100000 = [1.1646, 1.0881, 1.3380, 1.1206, 1.1153]
b_times_1000000 = [21.2837, 20.0075, 20.0632, 20.4115, 20.2848]

a_times_100 = [0.0000, 0.0000, 0.0000, 0.0000, 0.0010]
a_times_1000 = [0.0109, 0.0096, 0.0119, 0.0095, 0.0101]
a_times_10000 = [0.9208, 1.1623, 1.2870, 1.2569, 1.1836]
a_times_100000 = [135.5281, 122.2875, 141.2688, 165.9068, 130.2860]

a_mean_100 = np.mean(a_times_100)
a_mean_1000 = np.mean(a_times_1000)
a_mean_10000 = np.mean(a_times_10000)
a_mean_100000 = np.mean(a_times_100000)
a_mean_1000000 = math.pow(10, 2) * a_mean_100000

b_mean_100 = np.mean(b_times_100)
b_mean_1000 = np.mean(b_times_1000)
b_mean_10000 = np.mean(b_times_10000)
b_mean_100000 = np.mean(b_times_100000)
b_mean_1000000 = np.mean(b_times_1000000)

a_mean_times = []
a_mean_times.append( np.round_(a_mean_100, decimals = 4))
a_mean_times.append(np.round_(a_mean_1000 ,decimals= 4))
a_mean_times.append(np.round_(a_mean_10000 ,decimals= 4))
a_mean_times.append(np.round_(a_mean_100000 ,decimals= 4))
a_mean_times.append(np.round_(a_mean_1000000 ,decimals= 4))

b_mean_times = []
b_mean_times.append( np.round_(b_mean_100, decimals = 4))
b_mean_times.append(np.round_(b_mean_1000 ,decimals= 4))
b_mean_times.append(np.round_(b_mean_10000 ,decimals= 4))
b_mean_times.append(np.round_(b_mean_100000 ,decimals= 4))
b_mean_times.append(np.round_(b_mean_1000000 ,decimals= 4))


print(a_mean_times)
print(b_mean_times)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sizes, a_mean_times, label='Array Implementation')
plt.plot(sizes, b_mean_times, label='Heap Implementation')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Nodes')
plt.ylabel('Average Runtime (s)')
plt.title('Comparison of Array vs. Heap Implementations of Dijkstra\'s Algorithm')
plt.legend()
plt.grid(True)
plt.show()
```
